```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Label Encoding
data={'Difficulty':['Easy','medium','hard'],
     'test_score':[50,60,70]}
df=pd.DataFrame(data)
encoder=LabelEncoder()
df['Difficulty_en']=encoder.fit_transform(df['Difficulty'])
print(df)
cor=df['Difficulty_en'].corr(df['test_score'])
print("Correlation:" ,cor)
```

```python
data={'color':['Red','Blue','Pink'],
     'test_score':[50,60,70]}
df=pd.DataFrame(data)

df_e=pd.get_dummies(df,columns=['color'])
print(df_e)
cor=df_e['color_Blue'].corr(df_e['test_score'])
print("Correlation:" ,cor)
```
